```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,8,4,3,6,2,5])
ypoints = np.array([3,10,4,6,8,9,7])

plt.plot(xpoints,ypoints,marker = 'X',ms = 20,mec = 'r',mfc = 'hotpink') # o , *, (.) , (,) , X ,x , H ,h,+,P,s,D,d,p,V,^,>,<,1,2,3,4,|,_
plt.show() #ms is marker size here and mec is marker edge color abbreviation mfc is markerfaceccolor
#we cAN also use hexadecimal color values
#unsupported color names will give value error
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,10,4,6,7,5])

plt.plot(ypoints,'o:c',ls = "solid",linewidth = '13.4') #this is marker refernce value 'y' stands for abbreviation of cyan colour other colors are w,m,r,g,b,c,k,y
# 'o' stands for dotted line other params are '-',':','--','-.'
plt.show() # ls is linestyle
#linestyle can be written as ls.

#dotted can be written as :.

#dashed can be written as --.
#for none use " or ''
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

font1 = {'family' : 'san serif', 'color' : 'Blue' , 'size' : 20.0}
font2 = {'family' : 'serif' , 'color' : 'darkred' , 'size' : 13.6}

plt.plot(x1, y1, x2, y2)

plt.title("Sports Data",fontdict = font1,loc = "center")
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Averge Calories Burnt" , fontdict = font2)

plt.grid(color = 'green' , linestyle = '--' , linewidth = 0.5,axis = "both")
#specify which lines to display default is both

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.ylabel('depth')
plt.xlabel('temperature of water')
plt.scatter(x, y, c=colors , cmap = 'winter')
plt.colorbar()
plt.suptitle('Depth and Temperature Relation of Oceans')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([9,3,4,5])
mylabels = ["apples","bananas","pineapples","mangoes"] #throws a value error if the number of label != number of elements in array
myexplode = [0.2 , 0.3 ,0 ,0]
mycolors = ["hotpink" ,  "red" , "green" , "yellow"]

plt.pie(y , labels = mylabels , startangle = 900 , explode = myexplode , shadow = True , colors = mycolors) # if written colors as color it will give a type error
plt.legend(title = 'four fruits')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for a plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')

# plt.show() ensures the plot is displayed
plt.show()
```
