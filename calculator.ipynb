```python
import mpmath
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Error! Division by zero."

def calculator():
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    while True:
        choose = input("Enter choice(1/2/3/4): ")

        if choose in ['1', '2', '3', '4']:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))

            if choose == '1':
                print(f"{num1} + {num2} = {add(num1, num2)}")

            elif choose == '2':
                print(f"{num1} - {num2} = {subtract(num1, num2)}")

            elif choose == '3':
                print(f"{num1} * {num2} = {multiply(num1, num2)}")

            elif choose == '4':
                print(f"{num1} / {num2} = {divide(num1, num2)}")

            next_calculation = input("Do you want to perform another calculation ? (yes/no): ")
            if next_calculation.lower() != 'yes':
                break
        else:
            print("Invalid Input")

calculator()
```

```python
import tkinter as tk

class Calculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Calculator")
        self.root.geometry("400x600")
        
        self.expression = ""
        
        # Entry widget to display the input and result
        self.input_text = tk.StringVar()
        self.entry = tk.Entry(root, textvariable=self.input_text, font=('Arial', 24), bd=10, insertwidth=2, width=14, borderwidth=4)
        self.entry.grid(row=0, column=0, columnspan=4)
        
        # Buttons layout
        buttons = [
            '7', '8', '9', '/', 
            '4', '5', '6', '*', 
            '1', '2', '3', '-', 
            '0', '.', '=', '+'
        ]
        
        row_val = 1
        col_val = 0
        
        for button in buttons:
            self.create_button(button, row_val, col_val)
            col_val += 1
            if col_val > 3:
                col_val = 0
                row_val += 1
        
        self.clear_button = tk.Button(root, text='C', padx=40, pady=20, font=('Arial', 18), command=self.clear)
        self.clear_button.grid(row=row_val, column=0, columnspan=4)
    
    def create_button(self, value, row, col):
        button = tk.Button(self.root, text=value, padx=40, pady=20, font=('Arial', 18), command=lambda: self.click(value))
        button.grid(row=row, column=col)
    
    def click(self, value):
        if value == "=":
            try:
                result = str(eval(self.expression))
                self.input_text.set(result)
                self.expression = result
            except:
                self.input_text.set("Error")
                self.expression = ""
        else:
            self.expression += str(value)
            self.input_text.set(self.expression)
    
    def clear(self):
        self.expression = ""
        self.input_text.set("")

# Creating the main window
root = tk.Tk()
calculator = Calculator(root)
root.mainloop()
```
